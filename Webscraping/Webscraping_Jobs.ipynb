#### Function to Extract Job Title, Company, Job Summary, Location & Salary

```python
#function to extract title, company, summary, location and salary

def jobs_scrape(soup):

#   extract titles        
    title = soup.findAll(class_='jobtitle')
    for i in title:
        try:
            titles.append(i.text.strip())
        except:
            titles.append(np.nan)

#   extract companies
    company = soup.findAll(class_='company')
    for i in company:
        try:
            companies.append(i.text.strip())
        except:
            companies.append(np.nan)            

#   extract summaries
    summary = soup.findAll(class_='summary')
    for i in summary:
        try:
            summaries.append(i.text)
        except:
            summaries.append(np.nan)            
                       
#   extract locations
    loc = soup.findAll(class_='location')
    for i in loc:
        try:
            locations.append(i.text)
        except:
            locations.append(np.nan)

#   extract salaries
    results = soup.findAll('div', class_ = "result")
    for i in results:
        salary_table = i.find('td', class_= "snip")
        if salary_table:
            salaries.append(salary_table.find('nobr'))
        else:
            salaries.append(np.nan)
```

#### Initialize Lists for Storing Results

```python
titles = []
companies = []
summaries = []
locations = []
salaries = []
```

#### Search Items

```python
# specify name of city
city = 'New+York'

#specify job title
job = 'data+scientist'

#specify maximum number of results
max_results = 500
```

#### Main Loop

```python
increment = 10
for start in range(0, max_results, increment):
    URL = ('http://www.indeed.com/jobs?q='+job+'&l='+city+'&start='+str(start))
    r = requests.get(URL)
    soup = BeautifulSoup(r.content, "lxml")
    jobs_scrape(soup)

# confirm all lists have the same length
print 'list lengths', len(titles), len(companies), len(summaries), len(locations), \
len(salaries)
```

#### Create Dataframe

```python
df = pd.DataFrame([titles, companies, summaries, locations, salaries]).T
df.columns = ['title', 'company', 'summary', 'location', 'salary']

#Remove new line character from job summary text
df.summary = df.summary.apply(lambda x: re.sub('\n', '', x)) 
df
```
